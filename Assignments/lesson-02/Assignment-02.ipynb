#### 1. Random Choose Method to get optimal k* and *b

````python
from sklearn.datasets import load_boston
````

````python
data = load_boston()
````

````python
X, y = data['data'], data['target']
X_rm = X[:, 5]
````

````python
def price(x,r,b):
    return x*r+b
````

````python
import numpy as np
def loss(y,y_hat):
    e = np.array(y)-np.array(y_hat)
    return (e@e.T)/len(y)
````

````python
import random
trying_times = 2000
min_loss = float('inf')
best_k, best_b = None, None
for i in range(trying_times):
    k = random.random()*200-100
    b = random.random()*200-100
    price_by_random_k_and_b = [price(r, k, b) for r in X_rm]
    current_loss = loss(y, price_by_random_k_and_b)
    if current_loss < min_loss:
        min_loss = current_loss
        best_k, best_b = k, b
        print('when time is {},k is {},b is{},loss is{}'.format(i,best_k,best_b,min_loss))  
````

#### 2.Supervised Direction to get optimal k* and *b

````python
trying_times = 2000
min_loss = float('inf')
best_k = random.random() * 200 - 100
best_b = random.random() * 200 - 100
direction = [
    (+1, -1), 
    (+1, +1), 
    (-1, -1), 
    (-1, +1),
]
next_direction = random.choice(direction)
scalar = 0.1
for i in range(trying_times):
    k_direction, b_direction = next_direction  
    current_k, current_b = best_k + k_direction * scalar, best_b + b_direction * scalar
    price_by_k_and_b = [price(r, current_k, current_b) for r in X_rm]
    current_loss = loss(y, price_by_k_and_b)   
    if current_loss < min_loss: 
        min_loss = current_loss
        best_k, best_b = current_k, current_b
        next_direction = next_direction
        print('When time is : {}, get best_k: {} best_b: {}, and the loss is: {}'.format(i, best_k, best_b, min_loss))
    else:
        next_direction = random.choice(direction)
````

#### 3.Gradient Descent to get optimal k* and *b

````python
def partial_k(x, y, y_hat):
    n = len(y)

    gradient = 0
    
    for x_i, y_i, y_hat_i in zip(list(x), list(y), list(y_hat)):
        gradient += (y_i - y_hat_i) * x_i
    
    return -2 / n * gradient


def partial_b(x, y, y_hat):
    n = len(y)

    gradient = 0
    
    for y_i, y_hat_i in zip(list(y), list(y_hat)):
        gradient += (y_i - y_hat_i)
    
    return -2 / n * gradient
````

````python
trying_times = 20000
current_k = random.random()*200-100
current_b = random.random()*200-100
min_loss = float('inf')
learning_rate = 1e-04
for i in range(trying_times):
    price_k_b = [price(x,current_k,current_b) for x in X_rm]
    if loss(y,price_k_b) < min_loss:
        min_loss = loss(y,price_k_b)
        if i%50 == 0:
            print("When time is {},k is {},b is {},the loss is {}".format(i,current_k,current_b,min_loss))
    gradient_k = partial_k(X_rm, y, price_k_b)
    gradient_b = partial_b(X_rm, y, price_k_b)
    current_k = current_k + (-1 * gradient_k) * learning_rate
    current_b = current_b + (-1 * gradient_b) * learning_rate
````

#### 4. Try different Loss function and learning rate.

````python
def partial_k_2(x, y, y_hat):
    n = len(y)

    gradient = 0
    
    for x_i, y_i, y_hat_i in zip(list(x), list(y), list(y_hat)):
        if y_i<y_hat_i:
            gradient += x_i
        elif y_i ==y_hat_i:
            gradient += 0
        else:
            gradient += -x_i
    
    return gradient / n 


def partial_b_2(x, y, y_hat):
    n = len(y)

    gradient = 0
    
    for y_i, y_hat_i in zip(list(y), list(y_hat)):
        if y_i<y_hat_i:
            gradient += 1
        elif y_i ==y_hat_i:
            gradient += 0
        else:
            gradient += -1
    
    return gradient / n 
````

##### 3. Build the search agent

> Build the search agent based on the graph we build.

for example, when you run: 

```python

```
you need get the result: 

-> A -> B -> C -> ... -> 



````python
from collections import defaultdict
````

````python
station_net = defaultdict(list)
station_net.update(station_connection)
````

````python
def search(start,destination,graph,sort_candidate):
    routes = [[start]]
    seen = set()
    while routes:
        route = routes.pop(0)
        frontier = route[-1]
        if frontier in seen: continue
        for station in graph[frontier]:
            if station in seen: continue
            new_route= route + [station]
            routes.append(new_route)
            if station == destination: return new_route
        routes = sort_candidate(routes)
        seen.add(frontier)
````

````python
def transfer_stations_first(pathes): 
    return sorted(pathes, key=len)
````

````python
def pretty_print(route):
    return '->'.join(route)
````

````python
pretty_print(search('','',station_net,transfer_stations_first))
````
